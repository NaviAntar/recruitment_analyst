```python
# Mapping tipe pandas → BigQuery
dtype_mapping = {
    'object': 'STRING',
    'string': 'STRING',
    'Int64': 'INT64',
    'int64': 'INT64',
    'float64': 'FLOAT64',
    'boolean': 'BOOL',
    'bool': 'BOOL',
    'datetime64[ns]': 'DATE'
}

def generate_bq_schema(df):
    schema = []
    for col, dtype in df.dtypes.items():
        dtype_str = str(dtype)
        bq_type = dtype_mapping.get(dtype_str, 'STRING')
        schema.append({'name': col, 'type': bq_type})
    return schema
```

```python
# Definsiikan schema di datamart
table_schema_datamart = generate_bq_schema(data_before_cleaning)
table_schema_datamart
```

Merubah data frame dengan format KPI wide menjadi KPI Long sehingga analisa lebih bersih

```python
kpi_cols = [c for c in data_before_cleaning.columns if c.startswith("kpi_")]
```

```python
data_recruitment_long = data_before_cleaning.melt(
    id_vars=[
        "candidate_id",
        "position",
        "experience_years",
        "major",
        "applied_date"
    ],
    value_vars=kpi_cols,
    var_name="kpi_name",
    value_name="score"
)
```

```python
display(data_recruitment_long)
```

```python
table_schema_datamart1 = generate_bq_schema(data_recruitment_long)
table_schema_datamart1
```

```python
# Upload ke BigQuery
to_gbq(
    data_recruitment_long,
    destination_table = 'recruitment_analyst.data_long_after_cleaning',
    project_id = 'dqlab21-480410',
    if_exists = 'append',
    table_schema = table_schema_datamart1
)
```
